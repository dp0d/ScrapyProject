```python
import json
import pandas as pd
with open('coords.json','r',encoding = 'utf-8') as f:
    location_data = json.load(f)

df = pd.read_csv('esf_info.csv')
df = df.dropna()
df["province"] = df["province"].apply(lambda x: x.replace('直辖市', ''))

df["Full Name"] = df["province"].map(str) + df["city"]


df["total_price"] = df["total"].apply(lambda x: float(x.replace('万元', '')))

df['area_num'] = df["area"].apply(lambda x: float(x.replace('㎡', '')))

df['unit_pice'] = df['total_price'].map(float) / df['area_num'] * 10000



dic = dict(df.groupby('Full Name')['unit_pice'].mean())
# dic
```

```python
city_lis = list(set(list(df['Full Name'].values)))
city_lis = [x.replace('直辖市','') for x in city_lis]
for d in location_data:
    city = list(d.keys())[0]
    if city in city_lis:
        value = list(d.values())
        value[0].append(int(dic[city]))
        print((city,value[0]),',')
```
